```python
import numpy as np
import bnlearn
from IPython.display import display
```

```python
# Load alarm dataset and true model
df = bnlearn.import_example('alarm')
model_true = bnlearn.import_DAG('alarm')

display(df.head())
G = bnlearn.plot(model_true)
```

```python
model_learned = bnlearn.structure_learning.fit(df)

bnlearn.plot(model_learned)
```

```python
# Compare 2 models
bnlearn.compare_networks(model_true, model_learned, pos=G['pos'])
```

```python
# Fit parameter
DAG = bnlearn.import_DAG('alarm', CPD=False)
DAG_update = bnlearn.parameter_learning.fit(model_true, df)
```

```python
bnlearn.plot(DAG_update)
```
